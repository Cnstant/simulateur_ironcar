```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread('../ground/ground_1.JPG')
img = cv2.resize(img, (600,600))
```

```python
x0 = 250
y0 = 600
x1 = 600
y1 = 10
```

```python
origine = np.array((x0, y0))
fin = np.array((x1, y1))
```

```python
milieu = (origine + fin)/2
```

```python
vect = fin - origine
```

```python
longueur = np.sqrt(vect[0]**2 + vect[1]**2)
```

```python
vecteur_normal = np.array((-vect[1], vect[0]))
```

```python
vecteur_normal_unitaire = vecteur_normal/np.sqrt(vecteur_normal[0]**2 + vecteur_normal[1]**2)
```

```python
entre_axe = 100
```

```python
rayon_min = 20* (2*600**2 - (longueur/2)**2)/(2*600*(milieu[0]*vecteur_normal_unitaire[0]+milieu[1]*vecteur_normal_unitaire[1]))
```

```python
center  = milieu + 100 * vecteur_normal_unitaire
```

```python
center
```

```python
rayon = origine - center
```

```python
rayon
```

```python
rayon_valeur = np.sqrt(rayon[0]**2 + rayon[1]**2)
```

```python
rayon_valeur
```

```python
np.sqrt((600 - center[0])**2 + (600 - center[1])**2)
```

```python
plt.imshow(cv2.ellipse(img, (int(center[0]), int(center[1])) ,(int(rayon_valeur), int(rayon_valeur)), 0,0,360 ,(255,255,255), thickness=10))
plt.imshow(cv2.ellipse(img, (int(center[0]), int(center[1])) ,(int(rayon_valeur)-entre_axe, int(rayon_valeur)-entre_axe), 0,0,360 ,(255,255,255), thickness=10))
plt.imshow(cv2.ellipse(img, (int(center[0]), int(center[1])) ,(int(rayon_valeur)+entre_axe, int(rayon_valeur)+entre_axe), 0,0,360 ,(255,255,255), thickness=10))
plt.imshow(img)
```
